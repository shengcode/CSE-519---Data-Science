# insert useful calculated features 

logRank=log(rank) <br>
logReverseRank=log(101-rank) <br>
Y_year=2018-year <br>
logY_year=log(2018-year) <br>
logPopularity=log(popularity) <br>
FitY=logPopularity-logRank <br>
const_x all ONES=const_x <br>

```python
def insert_useful_features (df):
    newdf_1=df.assign(logRank=pd.Series(np.log(df['rank']))) # rank should be 1--100 
    newdf_2=newdf_1.assign(logReverseRank=pd.Series(np.log(101-newdf_1['rank']))) # 101-rank=1---100 
    newdf_3=newdf_2.assign(Y_year=pd.Series(2018-newdf_2['year'])) 
    newdf_4=newdf_3.assign(logY_year=pd.Series(np.log(2018-newdf_3['year'])))
    newdf_5=newdf_4.assign(logPopularity=pd.Series(np.log(newdf_4['popularity'])))
    newdf_6=newdf_5.assign(FitY=pd.Series(newdf_5['logPopularity']-newdf_5['logRank']))
    newdf_6['const_x']=1   
    return newdf_6
```

# model trying to fit is: Popularity=Rank^a * exp^(b*Y_year+c)<br>
logPopularity=a*logRank+b*Y_year+c (c is constant) <br>
X_train=[logRank, Y_year] <br>
Y_train=logPopularity <br>

```python
def model_one(df_train, df_test):
    X_train_1 = df_train[['logReverseRank']].values
    X_train_2=df_train[['Y_year']].values
    X_train=np.column_stack((X_train_1,X_train_2))
    print(X_train.shape)
    y_train=df_train['logPopularity'].values
    X_test_1 = df_test[['logReverseRank']].values
    X_test_2=df_test[['Y_year']].values
    X_test=np.column_stack((X_test_1,X_test_2))
    print(X_test.shape)
    y_test=df_test['logPopularity'].values 
    
    regr = linear_model.LinearRegression(fit_intercept=True) # with intercept 
    regr.fit(X_train, y_train)
    ytest_pred = regr.predict(X_test)
    ytrain_pred=regr.predict(X_train)

    print('Coefficients: \n', regr.coef_)
    print('Intercept when fit_intercept=True : {:.5f}'.format(regr.intercept_))
    print("Mean squared error for test case is: %.2f"% mean_squared_error(y_test, ytest_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_test, ytest_pred))
    print("Mean squared error for train case is: %.2f"% mean_squared_error(y_train, ytrain_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_train, ytrain_pred))

    df_train_mid=df_train.assign(predictedlogP=pd.Series(ytrain_pred))
    df_test_mid=df_test.assign(predictedlogP=pd.Series(ytest_pred))
    
    
    
    predict_train=df_train_mid.assign(predictedPopularity=pd.Series( np.exp(df_train_mid['predictedlogP'])))
    predict_test=df_test_mid.assign(predictedPopularity=pd.Series( np.exp(df_test_mid['predictedlogP'])))
    
    return predict_train, predict_test
   
    
    
```

```python
def plot_absolute_error_ofYear(predict_train, predict_test):
     # calculate absolute error of each year 
    predicted_test_year=predict_test.groupby('year')
    min_year=predict_test['year'].min()
    max_year=predict_test['year'].max()+1    
    
    error_year=[]
    for index, group in predicted_test_year:
        error= np.array(group['popularity']-group['predictedPopularity'])
        length=error.shape[0]
        error=np.abs(error)
        total_error=error.sum() 
        total_error=total_error/length
        error_year.append(total_error) 
    
    year = range(min_year, max_year)
    
    print("the mean absolute error of different years is: ")
    print(sum(error_year)/len(error_year))
    plt.figure(figsize=(20, 6), dpi=80)
    plt.plot(year,error_year)
    plt.xlabel("Year")
    plt.ylabel("Absolute Error")
    plt.title("for test case")
    plt.show()
    
    
    # PLOT BAR FIRUGE     
    plt.figure(figsize=(20, 6), dpi=80)
    plt.bar(year,error_year)
    plt.xlabel("Year",fontsize=18)
    plt.ylabel("Absolute Error",fontsize=18)
    plt.title("Absolute Error of Every Year",fontsize=18)
    plt.rc('xtick',labelsize=14)
    plt.rc('ytick',labelsize=14)
    plt.show()
    
    
    
```

```python
def plot_mean_square_errorofYear(predict_train, predict_test):
    predicted_test_year=predict_test.groupby('year')
    min_year=predict_test['year'].min()
    max_year=predict_test['year'].max()+1    
    
    error_year=[]
    for index, group in predicted_test_year:
        error= np.array(group['popularity']-group['predictedPopularity'])
        length=error.shape[0]
        error=np.square(error)        
        total_error=error.sum() 
        total_error=total_error/length
        error_year.append(total_error) 
    
    year = range(min_year, max_year)
    
    plt.figure(figsize=(20, 6), dpi=80)
    plt.plot(year,error_year)
    plt.xlabel("Year")
    plt.ylabel("Mean Square Error")
    plt.title("for test case")
    plt.show()
```

# popularity=rank*(2018-year)^a

```python
def model_two(df_train, df_test): 
    X_train = df_train[['logY_year']].values
    y_train=df_train['FitY'].values
    
    X_test = df_test[['logY_year']].values
    y_test=df_test['FitY'].values   
   
    
    regr = linear_model.LinearRegression() # without intercept 
    regr.fit(X_train, y_train)
    
    ytest_pred = regr.predict(X_test)
    ytrain_pred=regr.predict(X_train)

    print('Coefficients: \n', regr.coef_)
    
    print("Mean squared error for test case is: %.2f"% mean_squared_error(y_test, ytest_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_test, ytest_pred))
    print("Mean squared error for train case is: %.2f"% mean_squared_error(y_train, ytrain_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_train, ytrain_pred))

    df_train_mid=df_train.assign(predictedFitY=pd.Series(ytrain_pred))
    df_test_mid=df_test.assign(predictedFitY=pd.Series(ytest_pred))
    
    predict_train=df_train_mid.assign(predictedPopularity=pd.Series( np.exp(df_train_mid['predictedFitY']+df_train_mid['logRank'])))
    predict_test=df_test_mid.assign(predictedPopularity=pd.Series(np.exp(df_train_mid['predictedFitY']+df_train_mid['logRank'])))
    
    return predict_train, predict_test
```

# popularity=rank* exp(a*Y_year+c) c is constant 

```python
def model_three(df_train, df_test): 
    X_train = df_train[['Y_year']].values
    y_train=df_train['FitY'].values
    
    X_test = df_test[['Y_year']].values
    y_test=df_test['FitY'].values   
   
    
    regr = linear_model.LinearRegression(fit_intercept=True) # with intercept 
    regr.fit(X_train, y_train)
    ytest_pred = regr.predict(X_test)
    ytrain_pred=regr.predict(X_train)

    print('Coefficients: \n', regr.coef_)
    print('Intercept when fit_intercept=True : {:.5f}'.format(regr.intercept_))
    print("Mean squared error for test case is: %.2f"% mean_squared_error(y_test, ytest_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_test, ytest_pred))
    print("Mean squared error for train case is: %.2f"% mean_squared_error(y_train, ytrain_pred))
    print('Variance score for test case is: %.2f' % r2_score(y_train, ytrain_pred))

#     df_train.insert(loc=0, column='predictedFitY', value=ytrain_pred)
#     df_test.insert(loc=0, column='predictedFitY', value=ytest_pred)

    df_train_mid=df_train.assign(predictedFitY=pd.Series(ytrain_pred))
    df_test_mid=df_test.assign(predictedFitY=pd.Series(ytest_pred))
    
    predict_train=df_train_mid.assign(predictedPopularity=pd.Series( np.exp(df_train_mid['predictedFitY']+df_train_mid['logRank'])))
    predict_test=df_test_mid.assign(predictedPopularity=pd.Series(np.exp(df_train_mid['predictedFitY']+df_train_mid['logRank'])))
    
    return predict_train, predict_test
```

```python
def advanced_insert_feature(df, func = None, feature_name = None):
    if not feature_name:
        raise ('Empty feature')
    new_data = df.apply(lambda x: func(x, feature_name), axis = 1)
    df.insert(df.shape[1], feature_name, new_data)
    
def insert_feature(df, func = None, feature_name = None):
    if not feature_name:
        raise ('Empty feature')
    new_data = df.apply(func, axis = 1)
    df.insert(df.shape[1], feature_name, new_data)

def new_feature(row_data, feature_name = None):
    if not feature_name:
        raise ('Empty feature')
    songid = row_data['song']    
    if df_features[feature].loc[df_features[feature]['song'] == songid].shape[0] > 0:
        return 1
    else:
        return 0
```
