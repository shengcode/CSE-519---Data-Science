# Auxiliary Functions

```python
def print_statistics_error_info(df, model_name, popularity, predicted_popularity):
    print('Mean Absolute Error of' , model_name, 'is', mean_absolute_error(df[popularity], df[predicted_popularity]))
    print('Median Absolute Error of' , model_name, 'is', median_absolute_error(df[popularity], df[predicted_popularity]))
    print('Mean Squared Error of' , model_name, 'is', mean_squared_error(df[popularity], df[predicted_popularity]))
```

# Linear Regression Base Model
$
\text{popularity} = C_i\text{(reverse_rank)}^{C_r}(\text{weeks_on_chart})^{C_w}e^{C_t (\text{diff_year})}
$


$$
\begin{align*}
\log(\text{popularity}) =& \log C_i + C_r\log(\text{reverse_rank}) + C_w\log(\text{weeks_on_chart}) + {C_t (\text{diff_year})}
\end{align*}
$$

```python
def model_linear_regression(train = df_train, x_features = None, y_feature = None, intercept = True):
    x_train = train[x_features].values
    y_train = train[y_feature].values
    log_y_train = train[y_feature].apply(lambda x: 0 if x == 0 else np.log(x)).values
    regr = LinearRegression(fit_intercept=intercept)
    regr.fit(x_train, log_y_train)    
    return regr
```

```python
def predict_value_by_linear_regression(row_data, mode, jump_mode):
    feature = MODE[mode]
    jump = row_data['jump_time'] != 0
    x_val = row_data[lr_x_features].values
    val = None
    
    if mode == 'default':
        val = np.exp(lr_regr[mode][jump_mode].predict([x_val]))
    else:
        key = row_data[feature]
        if not JUMP_MODE[jump_mode]:
            val = np.exp(lr_regr[mode][jump_mode][key].predict([x_val]))
        else:
            val = np.exp(lr_regr[mode][jump_mode][jump][key].predict([x_val]))

    if val > 100:
        val = 100
    elif val < 0:
        val = 0
    else:
        val = val.round(decimals = 0)
    
    return val
```

# plot the difference of popularity  by year and by rank 

```python
def plot_percentage_of_over_under_perform_byYear(df,filename,title):
    year_group = df.groupby('year')
    year_array=[]
    sample_number=[]
    overperform_percent_array=[]
    underperform_percent_array=[]
    for idx, group in year_group:
        pop_array=group['popularity']
        pred_pop_array=group['pred_by_lr_r']
        total_sample=group.shape[0]
        overperform_percent=group.loc[group['pop_diff']>10].shape[0]/group.shape[0]
        underperform_percent=group.loc[group['pop_diff']<-10].shape[0]/group.shape[0]
        year_array.append(idx)
        overperform_percent_array.append(overperform_percent)
        underperform_percent_array.append(underperform_percent)
    plt.figure(figsize=(20,15))
    ax=plt.gca()
    ax.scatter(year_array, underperform_percent_array,marker='o',color='red',label="underperform percent")
    ax.scatter(year_array, overperform_percent_array, marker='o',color='green',label="overperform percent ")
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend()
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)
    plt.xlabel("year",fontsize=36)
    plt.ylabel("percent",fontsize=36)
#     plt.title(title,fontsize=36)
    plt.legend(loc='upper right', shadow=True, fontsize=30)
    plt.savefig(filename+'.png', bbox_inches='tight', pad_inches=0.0)
#     plt.show()
```

```python
def plot_percentage_of_over_under_perform_byRank(df,filename,title):
    rank_group= df.groupby('rank')
    rank_array=[]
    sample_number=[]
    overperform_percent_array=[]
    underperform_percent_array=[]
    for idx, group in rank_group:
        pop_array=group['popularity']
        pred_pop_array=group['pred_by_lr_r']
        total_sample=group.shape[0]
        overperform_percent=group.loc[group['pop_diff']>10].shape[0]/group.shape[0]
        underperform_percent=group.loc[group['pop_diff']<-10].shape[0]/group.shape[0]
        rank_array.append(idx)
        overperform_percent_array.append(overperform_percent)
        underperform_percent_array.append(underperform_percent)
    plt.figure(figsize=(20,15))
    ax=plt.gca()
    ax.scatter(rank_array, underperform_percent_array,marker='o',color='red',label="underperform percent")
    ax.scatter(rank_array, overperform_percent_array, marker='o',color='green',label="overperform percent ")
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend()
    plt.xticks(fontsize=30)
    plt.yticks(fontsize=30)
    plt.xlabel("rank",fontsize=36)
    plt.ylabel("percent",fontsize=36)
#     plt.title(title,fontsize=36)
    plt.legend(loc='upper right', shadow=True, fontsize=30)
    plt.savefig(filename+'.png', bbox_inches='tight', pad_inches=0.0)
#     plt.savefig('r.png', bbox_inches='tight', pad_inches=0.0)
#     plt.show() 
    
```
