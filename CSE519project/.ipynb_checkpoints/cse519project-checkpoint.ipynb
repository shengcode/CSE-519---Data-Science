```python
def compare(s1, s2):
    s1=s1.lower()
    s2=s2.lower()
    s1=s1.replace(' ', '')
    s2=s2.replace(' ', '')
    
    s3=s1.replace('&', '')
    s4=s2.replace('&', '')
    
    s5=s1.replace('&','and')
    s6=s2.replace('&','and')
    
    s7=s5.replace('\'s','s')
    s8=s6.replace('\'s','s')
    
    s9=s7.replace('-','')
    s10=s8.replace('-','')
    return s3== s4 or s9==s10 or (s9 in s10) or (s10 in s9)
```

```python
def intersect_n_times(n, i, arr):
    result= set(arr[i])
    for j in range(1,n+1):
        result.intersection_update(arr[i+j])
    return result 
```
